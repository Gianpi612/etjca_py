```python
import csv
import random

# ==============================================================================
# CONFIGURAZIONE PARAMETRI DI GENERAZIONE
# ==============================================================================
NUM_RECORDS = 3000  # Target incrementato a 3000 unità
FILENAME = 'dataset_analisi_formazione_avanzato.csv'

# Parametri di controllo qualità del dato
PROB_SBRIGATIVO = 0.03  # Bias di risposta casuale
PROB_COMPIACENTE = 0.03 # Bias di risposta affermativa sistematica
CHANCE_MISSING = 0.02   # Frequenza fisiologica di valori mancanti (NaN)

# ==============================================================================
# DEFINIZIONE ARCHETIPI UTENTE (TARGET DISOCCUPATI - TERZO SETTORE)
# ==============================================================================
# Ogni profilo modella una distribuzione demografica e comportamentale specifica.
# I pesi del genere riflettono i trend attuali del mercato del lavoro italiano.
# Pesi Genere: [Maschio, Femmina, Altro, Non Rispondere]
# Pesi Risposte Standard: [Sì, No, Non saprei/Ancora]
ARCHETIPI = {
    "Reskilling_3050": {
        "prob": 0.50, 
        "eta": "30-50",
        "peso_genere": [38, 59, 2, 1], # Prevalenza femminile (riqualificazione post-maternità/cura)
        "titoli": ["diploma", "laurea_triennale", "triennale_master", "laurea_magistrale"],
        "pesi_titoli": [45, 35, 10, 10],
        "pesi_default": [75, 15, 10]
    },
    "Junior_EntryLevel": {
        "prob": 0.30, 
        "eta": "18-30",
        "peso_genere": [48, 48, 3, 1], # Sostanziale parità demografica tra i giovani
        "titoli": ["diploma", "laurea_triennale"],
        "pesi_titoli": [70, 30],
        "pesi_default": [60, 25, 15]
    },
    "Over50_Reinvention": {
        "prob": 0.20, 
        "eta": "50+",
        "peso_genere": [55, 42, 2, 1], # Prevalenza maschile (re-skilling da settori industriali)
        "titoli": ["diploma", "laurea_triennale"],
        "pesi_titoli": [85, 15],
        "pesi_default": [55, 30, 15]
    }
}

# ==============================================================================
# DEFINIZIONE DOMANDE E OPZIONI DI RISPOSTA
# ==============================================================================
DOMANDE = [
    {"id": "q1", "text": "Genere", "opts": ["maschio", "femmina", "altro", "non_rispondere"]},
    {"id": "q2", "text": "Area Residenza", "opts": ["sud", "centro", "nord"]},
    {"id": "q3", "text": "Fascia Età", "opts": ["18-30", "30-50", "50+"]},
    {"id": "q_titolo", "text": "Titolo Studio", "opts": ["diploma", "laurea_triennale", "triennale_master", "laurea_magistrale", "post_magistrale"]},
    {"id": "q4", "text": "Chiarezza Professionale", "opts": ["si", "no", "non_saprei"]},
    {"id": "q5", "text": "Punti di Forza", "opts": ["si", "no", "non_saprei"]},
    {"id": "q6", "text": "Colmare Lacune", "opts": ["si", "no", "non_ancora"]},
    {"id": "q7", "text": "Rispetto Valori", "opts": ["si", "no", "non_saprei"]},
    {"id": "q8", "text": "Obiettivi 6 mesi", "opts": ["si", "no", "non_saprei"]},
    {"id": "q9", "text": "Competenze Tecniche", "opts": ["si", "no", "non_saprei"]},
    {"id": "q10", "text": "Soft Skills", "opts": ["si", "no", "non_saprei"]},
    {"id": "q11", "text": "Investimento Nuove Skill", "opts": ["si", "no", "non_ancora"]},
    {"id": "q12", "text": "Esplorazione Settore", "opts": ["si", "no", "non_ancora"]},
    {"id": "q13", "text": "Utilità Percorso HR", "opts": ["si", "no", "non_saprei"]},
    {"id": "q14", "text": "Monitoraggio Lavoro", "opts": ["si", "no", "non_ancora"]},
    {"id": "q15", "text": "Lavoro Trovato", "opts": ["si", "no", "non_ancora"]}
]

# ==============================================================================
# LOGICA DI GENERAZIONE RECORD
# ==============================================================================
def genera_record():
    r = random.random()
    record = {}
    
    # Selezione del profilo basata sulle probabilità degli archetipi e bias
    if r < PROB_SBRIGATIVO:
        tipo = "Sbrigativo"
    elif r < (PROB_SBRIGATIVO + PROB_COMPIACENTE):
        tipo = "Compiacente"
    else:
        tipo = random.choices(list(ARCHETIPI.keys()), weights=[a['prob'] for a in ARCHETIPI.values()])[0]

    for dom in DOMANDE:
        d_id = dom['id']
        opzioni = dom['opts']
        
        # 1. Gestione Profili Distorti
        if tipo == "Sbrigativo":
            valore = random.choice(opzioni)
        elif tipo == "Compiacente":
            valore = "si" if "si" in opzioni else random.choice(opzioni)
        
        # 2. Gestione Profili Analitici (Archetipi)
        else:
            arc = ARCHETIPI[tipo]
            if d_id == 'q1': 
                # Applicazione del peso di genere specifico per l'archetipo
                valore = random.choices(opzioni, weights=arc['peso_genere'])[0]
            elif d_id == 'q2': 
                valore = random.choices(opzioni, weights=[40, 25, 35])[0] # Distribuzione geografica
            elif d_id == 'q3': 
                valore = arc['eta']
            elif d_id == 'q_titolo':
                valore = random.choices(arc['titoli'], weights=arc['pesi_titoli'])[0]
            else:
                # Correlazione: Maggiore richiesta di supporto HR (q13) per profili senior
                pesi = [95, 2, 3] if (d_id == 'q13' and tipo == "Over50_Reinvention") else arc['pesi_default']
                valore = random.choices(opzioni, weights=pesi)[0]

        # 3. Introduzione Valori Mancanti (per simulare errori di data-entry o abbandoni)
        if d_id not in ['q1', 'q2', 'q3'] and random.random() < CHANCE_MISSING:
            record[d_id] = ""
        else:
            record[d_id] = valore
            
    return record

# ==============================================================================
# ESECUZIONE E SCRITTURA FILE
# ==============================================================================
with open(FILENAME, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = [d['id'] for d in DOMANDE]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    
    for _ in range(NUM_RECORDS):
        writer.writerow(genera_record())

print(f"REPORT DI GENERAZIONE DATASET:")
print(f"- File generato: {FILENAME}")
print(f"- Record totali: {NUM_RECORDS}")
print(f"- Clusterizzazione: Genere pesato per fascia d'età e profilo professionale.")
print("-" * 50)
```
